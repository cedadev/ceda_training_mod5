## Solution 1.2 : Plot layout and saving

#### Part 1

```python
from matplotlib import pyplot as plt

x = np.linspace(-5, 5, 40)
y = x**2
fig, axis = plt.subplots(1)
axis.plot(x, y, label='$y_1 = x^2$')
axis.plot(x, x + 6, label='$y_2 = x + 6$')
axis.legend(loc='upper center')
axis.set_xlabel('$x$')
axis.set_ylabel('$y$')
fig.savefig('data/ex_1.2.pdf')
```

#### Part 2

```python
fig, axes = plt.subplots(1, 2)

# [(xlim, ylim), ...]
regions = [((-3, -1), (0, 5)),
           ((2, 4), (5, 10))]

for axis, (xlim, ylim) in zip(axes, regions):
    axis.plot(x, y)
    axis.plot(x, x + 6)
    axis.set_xlim(xlim)
    axis.set_ylim(ylim)
    axis.set_xlabel('$x$')
    axis.set_ylabel('$y$')
    axis.grid()
    
fig.tight_layout()
```

The plots indicate intersections at $x=-2$ and $x=3$, consistent with $y = (x + 2)(x - 3) = x^2 - x - 6$.

## Solution 1.3 : Plotting 2D data

```python
# From the Exercise
from scipy.stats import norm

def f(a, x):
    y = norm(loc=a).pdf(x)
    Y = np.reshape(y, (1, len(y)))
    return Y.T * Y

x = np.linspace(-4, 4, 50)
Z = f(-1, x) + f(1, x)
```

#### Part 1

```python
plt.imshow(Z)
```

#### Part 2

```python
fig, ax = plt.subplots()
X, Y = np.meshgrid(x, x)
p = ax.contour(X, Y, Z, colors='k')
ax.clabel(p)
```

## Solution 2.2 : trapezoidal integration

#### Part 1

```python
x = np.linspace(0, 3, 10)
y = x**2
```

#### Part 2

```python
y[1:] + y[:-1]
```

#### Part 3

```python
def trapz(x, y):
    dx = x[1:] - x[:-1]
    dy = y[1:] + y[:-1]
    return np.sum(dx * dy) / 2
```

#### Part 4

```python
trapz(x, y)
```

#### Part 5 (extension)

```python
# Solution 2.2.5
import scipy.integrate

scipy.integrate.trapz(y, x)
```

## Solution 3 : Creating NetCDF

#### Part 1

```python
from netcdftime import num2date, date2num
from datetime import datetime, timedelta

dates = []
units = 'seconds since 2013-03-19 00:00:00'
dt = datetime(2013, 3, 19, 0, 0, 0)
while dt < datetime(2013, 3, 19, 12, 0, 0):
    dt = dt + timedelta(minutes=5)
    dates.append(dt)
    
times = date2num(dates, units=units)
print num2date(times[:5], units)
print num2date(times[-5:], units)
```

#### Part 2

```python
from netCDF4 import Dataset

# 1.
d = Dataset('data/ex5.nc', 'w', format='NETCDF4_CLASSIC')
time_d = d.createDimension('time', len(times))
time_v = d.createVariable('time', np.int32, ('time', ))
o3_v = d.createVariable('o3', np.float64, ('time', ))
```

```python
# 2.
o3_v[:] = sin(np.linspace(0, pi, len(times))) * 300
plot(o3_v)
```

```python
# 3.
o3_v[:] = o3_v + np.random.normal(size=len(times)) * 10
plot(o3_v)
```

## Solution 4.1 : Converting PP files to NetCDF

#### Part 1

```python
import cf
from glob import glob

pp_files = glob('data/aatzja.pm90*.pp')
datasets = [cf.read(f) for f in pp_files]
fields = [d.select('cloud_area_fraction') for d in datasets]
```

#### Part 2

```python
cloud_agg = cf.aggregate(fields)
print cloud_agg
```
